```python
import numpy as np
from scipy.integrate import quad
import matplotlib.pyplot as plt
```

```python
#QUESTÃO A
def f(x):
    return -x**2 + 3

def g(x):
    return x**2 + 1

a = -1
b = 1

area = quad(lambda x: f(x) - g(x), a, b)[0]

x = np.linspace(a, b, num=100)
y1 = f(x)
y2 = g(x)

plt.plot(x, y1, label='Curva Superior')
plt.plot(x, y2, label='Curva Inferior')
plt.fill_between(x, y1, y2, color='gray', alpha=0.5, label='Área')

plt.xlabel('X')
plt.ylabel('Y')
plt.title(f'Área entre Curvas: {area:.2f}')
plt.legend()
plt.grid()
plt.show()
```
