```python
import matplotlib.pyplot as plt
import numpy as np


def f(x):
    return x**2

x = np.linspace(0, 1, 1000)
plt.plot(x, f(x))
plt.axhline(color="black")
plt.fill_between(x, f(x), where=[(x > 0) and (x < 1) for x in x])
plt.show()
```

```python
import matplotlib.pyplot as plt
from numpy import arange, sin, cos, pi
import math

def f(x):
    return cos(x)

x = np.linspace(0, pi/2, 1000)
plt.plot(x, f(x))
plt.axhline(color="black")
plt.fill_between(x, f(x), where=[(x > 0) and (x < pi/2) for x in x])
plt.show()
```
