```python

import numpy as np
import matplotlib.pyplot as plt

def plot_square(ax, vertices, color='blue', title="Original"):
    """
    Função para plotar o quadrado.
    """
    # Adiciona o primeiro ponto ao final para fechar o quadrado
    vertices_ciclo = np.vstack([vertices, vertices[0]])
    ax.plot(vertices_ciclo[:, 0], vertices_ciclo[:, 1], color=color)
    ax.set_xlim(-12, 15)
    ax.set_ylim(-15, 20)
    ax.set_title(title)
    ax.set_aspect('equal')

# Vértices originais do quadrado
H = np.array([[1,1],[2,1],[2,3],[3,3],[3,1],[4,1],[4,6],[3,6],[3,4],[2,4],[2,6],[1,6]])
A_T=([[2,1],[1,2]])
T_O=np.dot(H,A_T)

#Trasnformação 1
A_T1=([[1,0],[0,2]])
T_1=np.dot(H,A_T1)

#Trasnformação 2
A_T2=([[1,1],[0,1]])
T_2=np.dot(H,A_T2)

#Trasnformação 3
A_T3=([[1,1],[0,2]]);
T_3=np.dot(H,A_T3)

#Trasnformação 4
A_T4=([[0,-1],[1,0]]);
T_4=np.dot(H,A_T4)

#Trasnformação 5
A_T5_1=([[-2,0],[0,-2]]);
T_5_1=np.dot(H,A_T5_1) #a = -2
A_T5_2=([[1/2,0],[0,1/2]]);
T_5_2=np.dot(H,A_T5_2) #a = 1/2

# Plota os resultados
fig, axs = plt.subplots(2, 4, figsize=(12, 8))

plot_square(axs[0,0], H,  title="Original")
plot_square(axs[0,1], T_O, title="transformação 0")
plot_square(axs[0,2], T_1, title="transformação 1")
plot_square(axs[0,3], T_2, title="transformação 2")
plot_square(axs[1,0], T_3, title="transformação 3")
plot_square(axs[1,1], T_4, title="transformação 4")
plot_square(axs[1,2], T_5_1, title="transformação 5.1")
plot_square(axs[1,3], T_5_2, title="transformação 5.2")
plt.show()

plt.show()
```

#**Explicação das Transformações Feitas**

0. A Letra H foi aumentada e inclinada para o lado direito.

1. A Letra H aumentou sua altura.

2. A Letra H inclinada e puxada para cima.

3. A Letra H que foi inclinada e puxada para cima aumentou de tamanho.

4. A Letra H foi inclinado á 90°.

5. A Letra H quanto a = 2, se torna maior duaz vezes maior e está abaixo de 0. Já quando a = 1/2 o tamanho original é diminuido pela metade e fica acima de 0.


#**Transformações Feitas**

1.   $T_1(x,y)=(x,2y)$, $$
A_{T_1}=\begin{bmatrix}
1 & 0 \\
0 & 2 \\
\end{bmatrix}
$$

2.   $T_2(x,y)=(x+y,y)$, $$
A_{T_2}=\begin{bmatrix}
1 & 1 \\
0 & 1 \\
\end{bmatrix}
$$

3. Utilizando a transformação $T_1$ e $T_2$ dos exercícios anteriores e fazendo uma composição das funções:
$T_3(x,y)=T_1(x,y)*T_2(x,y)$
$$A_T=A_{T_1}A_{T_2}=\begin{bmatrix}
1 & 0 \\
0 & 2 \\
\end{bmatrix}\begin{bmatrix}
1 & 1 \\
0 & 1 \\
\end{bmatrix}
$$

4. $T_4=(x\cos(\theta)-y\sin(θ),x\sin(\theta)+y\cos(θ)$ rotaçaõ de angulo θ ($0\leq θ\leq 2π$), para $θ={\pi}/2$
$$A_{T_4}=\begin{bmatrix}
cosθ & -senθ \\
senθ & cosθ \\
\end{bmatrix}
$$

5.$T_5(x,y)=α(x,y)$, Para ampliar ou diminuir a matrix
$$A_{T_5}=α\begin{bmatrix}
1 & 0 \\
0 & 1 \\
\end{bmatrix}$$
para $α = -2$ e $α=1/2$
