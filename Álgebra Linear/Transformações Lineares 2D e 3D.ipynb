```python
#Noemi Soares Gonçalves da Silva | RA:00000851329
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
import numpy as np

# QUADRADO
origem_quadrado = [1, 1, 0]

vertices_quadrado = np.array([[origem_quadrado[0], origem_quadrado[1], origem_quadrado[2]],
                              [origem_quadrado[0] + 1, origem_quadrado[1], origem_quadrado[2]],
                              [origem_quadrado[0] + 1, origem_quadrado[1] + 1, origem_quadrado[2]],
                              [origem_quadrado[0], origem_quadrado[1] + 1, origem_quadrado[2]]])

fig_quadrado, ax_quadrado = plt.subplots()
quadrado = plt.Polygon(vertices_quadrado[:, :2], closed=True, edgecolor='r', facecolor='cyan', alpha=0.25)
ax_quadrado.add_patch(quadrado)

ax_quadrado.set_xlim(-2, 2)
ax_quadrado.set_ylim(-2, 2)

ax_quadrado.set_xlabel('X')
ax_quadrado.set_ylabel('Y')
ax_quadrado.set_title('Quadrado 2D')
ax_quadrado.set_aspect('equal', 'box')

plt.show()

# CUBO
origem_cubo = [-1, -1, -1]

vertices_cubo = np.array([[origem_cubo[0], origem_cubo[1], origem_cubo[2]],
                          [origem_cubo[0] + 1, origem_cubo[1], origem_cubo[2]],
                          [origem_cubo[0] + 1, origem_cubo[1] + 1, origem_cubo[2]],
                          [origem_cubo[0], origem_cubo[1] + 1, origem_cubo[2]],
                          [origem_cubo[0], origem_cubo[1], origem_cubo[2] + 1],
                          [origem_cubo[0] + 1, origem_cubo[1], origem_cubo[2] + 1],
                          [origem_cubo[0] + 1, origem_cubo[1] + 1, origem_cubo[2] + 1],
                          [origem_cubo[0], origem_cubo[1] + 1, origem_cubo[2] + 1]])

faces_cubo = [[vertices_cubo[j] for j in [0, 1, 2, 3]],
              [vertices_cubo[j] for j in [4, 5, 6, 7]],
              [vertices_cubo[j] for j in [0, 1, 5, 4]],
              [vertices_cubo[j] for j in [2, 3, 7, 6]],
              [vertices_cubo[j] for j in [0, 3, 7, 4]],
              [vertices_cubo[j] for j in [1, 2, 6, 5]]]

fig_cubo = plt.figure()
ax_cubo = fig_cubo.add_subplot(111, projection='3d')
ax_cubo.add_collection3d(Poly3DCollection(faces_cubo, facecolors='cyan', linewidths=1, edgecolors='r', alpha=.25))

ax_cubo.set_xlim([-2, 2])
ax_cubo.set_ylim([-2, 2])
ax_cubo.set_zlim([-2, 2])

ax_cubo.set_xlabel('X')
ax_cubo.set_ylabel('Y')
ax_cubo.set_zlabel('Z')
ax_cubo.set_title('Cubo 3D')

plt.show()
```
